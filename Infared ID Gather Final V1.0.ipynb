```python
import cv2
import os
from pandas import *
import requests
import numpy as np
import json
import urllib
import pandas as pd
import openpyxl
import urllib.request
import urllib.error
```

```python
# Code for both generating manuscript-parchment-fragment format and gathering the identifiers.

fragments_data_path = 'fragmentsData'
output_file_path = 'output.txt'
output_file_path_2 = 'output2.txt'


for folder_name in os.listdir(os.path.join(fragments_data_path, 'DSS_Joins')):
    # Check if the item is a folder
    if os.path.isdir(os.path.join(fragments_data_path, 'DSS_Joins', folder_name)):
        # Extract manuscript_name, plate, and fragment from the folder name
        manuscript_name = folder_name

        # Iterate through the images inside the manuscript folder
        for image_name in os.listdir(os.path.join(fragments_data_path, 'DSS_Joins', folder_name)):
            # Check if the item is a file
            if os.path.isfile(os.path.join(fragments_data_path, 'DSS_Joins', folder_name, image_name)):
                # Extract the manuscript, plate, and fragment information from the image name
                image_parts = image_name.split('-')
                manuscript = manuscript_name
                #print(image_parts[0])   # Testing
                if image_parts[0][0]=="P":
                    plate = image_parts[0][1:]  # Remove the "P" prefix
                    #print(plate)   # Testing
                else:
                    plate = image_parts[0]
                    #print(plate)   # Testing
                fragment = image_parts[1][2:].lstrip('0')  # Remove the "Fg" prefix and preceding zeros


                # Construct the row for the txt file
                row = f'{manuscript}-{plate}-{fragment}.json'

                # Append the row to the output file
                with open(output_file_path, 'a') as f:
                    f.write(row + '\n')

print('Done')



def extract_link_from_json(json_url, condition):
    response = requests.get(json_url)
    data = json.loads(response.text)

    sequences = data['sequences']
    for sequence in sequences:
        canvases = sequence['canvases']
        for canvas in canvases:
            on = canvas['images'][0]['on']
            if condition in on:
                link = canvas['images'][0]['resource']['@id']
                return link

    return None



json_files_url = 'https://dss.digitalbibleonline.org/manifests/all/'
with open(output_file_path, 'r') as f:
    for line in f:
        line = line.rstrip()
        print(line)
        json_file_url = f"{json_files_url}{line}"
        json_file_url = json_file_url.replace(" ","%20",-1)
        print(json_file_url)
        link = json_file_url
        try:
            f = urllib.request.urlopen(link)
            myfile = f.read()
            #find3 = myfile.find("https://iaa.iiifhosting.com/iiif/".encode()) + 33
            #find4 = myfile.find("https://iaa.iiifhosting.com/iiif/".encode(), find3) + 33
            #find5 = myfile.find("https://iaa.iiifhosting.com/iiif/".encode(), find4) + 33
            find3 = myfile.find("Near Infra-Red (NIR) -Recto".encode()) + 30
            find4 = myfile.find("Near Infra-Red (NIR) -Recto".encode(), find3) + 30
            if find4 - 30 != -1:
                find = myfile.find("https://iaa.iiifhosting.com/iiif/".encode(), find4)
                find2 = myfile.find("/full/500,/0/default.jpg".encode(), find4)
                with open(output_file_path_2, 'a') as n:
                    n.write(myfile[find+33:find2].decode() + '\n')
            elif find3 - 30 != -1:
                find = myfile.find("https://iaa.iiifhosting.com/iiif/".encode(), find3)
                find2 = myfile.find("/full/500,/0/default.jpg".encode(), find3)
                with open(output_file_path_2, 'a') as n:
                    n.write(myfile[find+33:find2].decode() + '\n')
            else:
                print("error")
                with open(output_file_path_2, 'a') as n:
                    n.write("error" + '\n')                            
        except:
            print("error")
            with open(output_file_path_2, 'a') as n:
                n.write("error" + '\n')


link = extract_link_from_json(json_file_url, "Full Spectrum Color-Recto")
start_index = link.find('/iiif/') + 6
end_index = link.find('/full')
identifier = link[start_index:end_index]

identifier = 'd44af9bfc22fe5f4c285f79d28d4f95d6d046ac27f9c82e0eb2245a5801fc302'
```
